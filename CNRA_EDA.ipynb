Some brief takeaways:
- A lot of data - data goes back pretty far and there are a lot of different stations
- I was using stn_id and site_code interchangeably because it looked like they matched up, but maybe not?
- some of the stations, when plotted over time, don't have variation. 
- Some of the data that does have variation, doesn't have A LOT of variation (ie only increasing by 0.1) 
- There are ~ 4000 stations with data that has atleast one other wlm_rpe measurement. However probably need to implement a better way to see which wells actually have seasonality. 
- using the merged dataset may not be needed - only the measurements dataset has the target variable.
